```python
import pandas as pd
import csv
import numpy as np
from scipy.stats import ttest_ind
```

```python
info = pd.read_csv('info.csv')
data = pd.read_csv('data.csv')
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 2000)
match_id = np.unique(data['File No'])
```

```python
match_data = []
final_df = pd.DataFrame()

for match in match_id:
    match_df = data.loc[data['File No'] == match]
        
        
    team1 = info[info['File No']==match]['Team 1'].values[0]
    team2 = info[info['File No']==match]['Team 2'].values[0]
    toss_winner = info[info['File No']==match]['Toss Winner'].values[0]
    toss_decision = info[info['File No']==match]['Toss Decision'].values[0]    
    
    if(toss_winner == team1 and toss_decision == 'bat'):
        inning_played_by_toss_winner = 1
    elif(toss_winner == team1 and toss_decision != 'bat'):
        inning_played_by_toss_winner = 2
    
    if(toss_winner == team2 and toss_decision == 'bat'):
        inning_played_by_toss_winner = 1
    elif(toss_winner == team2 and toss_decision != 'bat'):
        inning_played_by_toss_winner = 2
        
        
    if(inning_played_by_toss_winner == 1):
        toss_winner_df = match_df.loc[match_df['Innings'] == 1]
        toss_loser_df = match_df.loc[match_df['Innings'] == 2]
    else:
        toss_winner_df = match_df.loc[match_df['Innings'] == 2]
        toss_loser_df = match_df.loc[match_df['Innings'] == 1]
    
    toss_winner_max = max(toss_winner_df['Runs'])
    toss_loser_max = max(toss_loser_df['Runs'])
    
    match_data.append({'File No': match, 'Toss Winner Max Runs': toss_winner_max, 'Toss Loser Max Runs': toss_loser_max})

final_df = final_df.append(match_data)
```

```python
mean1 = final_df['Toss Winner Max Runs'].mean()
mean2 = final_df['Toss Loser Max Runs'].mean()
```

```python
# Hypothesis 3
ttest_ind(final_df['Toss Winner Max Runs'], final_df['Toss Loser Max Runs'], nan_policy='omit')
```
